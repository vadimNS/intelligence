**Задача 1:** Який відсоток виживання серед чоловіків / жінок? 

````python
import pandas as pd

dataset = pd.read_csv('titanic.csv')

survival_percentages = dataset.groupby('Sex')['Survived'].mean() * 100

print("Відсоток виживших серед чоловіків та жінок:")
print(survival_percentages)
````

**Задача 4:** Назвіть найпопулярніше **жіноче** ім'я на кораблі. 

````python
import pandas as pd

dataset = pd.read_csv('titanic.csv')


female_names = dataset[(dataset['Sex'] == 'female') & (dataset['Name'].str.contains('Miss'))]


female_names['First Name'] = female_names['Name'].apply(lambda x: x.split(' ')[2])


popular_female_name = female_names['First Name'].value_counts().idxmax()

print("Найпопулярніше жіноче ім'я на кораблі через тег Miss: ", popular_female_name)
````

**Задача 5:** Назвіть найпопулярніше **чоловіче** ім'я на кораблі. 

````python
import pandas as pd
dataset = pd.read_csv('titanic.csv')
male_names = dataset[dataset['Sex'] == 'male']['Name']
first_names = male_names.apply(lambda x: x.split(',')[1].split()[1])
name_counts = first_names.value_counts()
most_popular_male_name = name_counts.idxmax()
print("Найпопулярніше чоловіче ім'я на кораблі:", most_popular_male_name)
````

**Задача 6:** Побудуйте діаграми співвідношення загиблих та виживших за інформацією про те, квитки якого класу були у пасажирів.
image.png

````python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('titanic.csv')

survived_by_class = dataset.groupby('Pclass')['Survived'].value_counts().unstack()

survived_by_class.plot(kind='bar', stacked=False, color=['red', 'green'])

plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.legend(['Dead', 'Survived'])

plt.show()

dataset = pd.read_csv('titanic.csv')

passenger_count_by_class = dataset['Pclass'].value_counts().sort_index()

plt.bar(passenger_count_by_class.index, passenger_count_by_class.values, color='skyblue')

plt.title('Passenger Count by Ticket Class')
plt.xlabel('Ticket Class')
plt.ylabel('Number of Passengers')
plt.xticks(passenger_count_by_class.index, ['1st Class', '2nd Class', '3rd Class'])

plt.show()
````

**Прокоментуйте побудовані графіки**:


1.   Цей графік відображає кількість виживших та загиблих пасажирів за класом квитків.
Кожний стовпець представляє один клас квитків (1-й, 2-й, 3-й), а висота стовпця відповідає загальній кількості пасажирів цього класу.
Стовпці розділені на дві частини: червоний - загиблі, зелений - виживших.
2.   Цей графік відображає загальну кількість пасажирів за кожним класом квитків.
Кожний стовпець представляє один клас квитків (1-й, 2-й, 3-й), а висота стовпця відповідає загальній кількості пасажирів цього класу.

**Задача 8** Побудуйте наступні діаграми розподілів:

image.png

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv('titanic.csv')

plt.figure(figsize=(8, 6))
sns.histplot(data=dataset[dataset['Survived'] == 0], x='Age', bins=20, color='red', kde=True)
plt.title('Frequency of Non-Survivors')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 6))
sns.histplot(data=dataset[dataset['Survived'] == 1], x='Age', bins=20, color='green', kde=True)
plt.title('Frequency of Survivors')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 8))
sns.histplot(data=dataset, x='Age', hue='Survived', multiple='stack', bins=20, palette=['red', 'green'])
plt.title('Dead vs Survived by Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(['Dead', 'Survived'])
plt.show()
````

1. Розподіл частоти для пасажирів, які не вижили (Survived = 0):
Гістограма показує розподіл віку серед пасажирів, які не вижили.
За допомогою цієї візуалізації можна зрозуміти, що серед пасажирів, які не вижили, було багато молодих людей, але також є певна кількість пасажирів середнього та старшого віку.

2. Розподіл частоти для пасажирів, які вижили (Survived = 1):
Гістограма відображає розподіл віку серед пасажирів, які вижили.
З цієї візуалізації видно, що серед виживших було більше молодих та середнього віку пасажирів, порівняно зі старшими.

3. Двовимірний розподіл віку між вижившими та загиблими пасажирами:
Ця гістограма надає двовимірний розподіл віку серед пасажирів, виживших та загиблих.
Вона показує, що серед виживших було більше молодих пасажирів, в той час як серед загиблих було більше старших пасажирів.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv('titanic.csv')

plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', data=dataset, palette='viridis')
plt.title('Number of Passengers Boarded')
plt.xlabel('Embarked')
plt.ylabel('Number of Passengers')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', hue='Sex', data=dataset, palette='Set2')
plt.title('Embarked Splited Female-Male')
plt.xlabel('Embarked')
plt.ylabel('Number of Passengers')
plt.legend(title='Sex')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', hue='Survived', data=dataset, palette='Set1')
plt.title('Embarked vs Survived')
plt.xlabel('Embarked')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', loc='upper right')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', hue='Pclass', data=dataset, palette='Set3')
plt.title('Embarked vs Pclass')
plt.xlabel('Embarked')
plt.ylabel('Number of Passengers')
plt.legend(title='Pclass', loc='upper right')
plt.show()
````
