**Завдання 4:** Як ви могли помітити, у нашому датесеті досить багато пропущених значень, зокрема у атрибута віку пасажирів. Робота з такими даними впливае як на результат моделювання так і на загалом уявлення цілісної картини про дані. Тож для того, що позбавитися пропусків ми використаємо заповнення пропущених значень медіаною за групами.



1.   Напишіть функцію, що на вхід приймає датафрейм, назву колонки з пропущеними значеннями та список групуючих колонок та повертає датафрейм з заповненими пропущеними значеннями
2.   Наведіть приклад роботи функції та зробіть перевірку на пропущені значення (дивись **завдання 2**)



````python
def fill_missing_with_median(df, column_name, groupby_columns):
    df_filled = df.copy()
    df_filled[column_name] = df_filled.groupby(groupby_columns)[column_name].transform(lambda x: x.fillna(x.median()))
    return df_filled

data = pd.read_csv('titanic.csv')


df_filled = fill_missing_with_median(data, 'Age', ['Sex', 'Pclass'])

print(df_filled.isnull().sum())
````

Ваші висновки: всі пропущені значення у колонці "Age" були заповнені за допомогою функції fill_missing_with_median

 
***Завдання 5:*** В моделях машинного навчання інснує велика проблема опрацювання безперервних змінних. Потрібно перетворити безперервні значення в категоріальні.

***Завдання 6:*** Так як змінна ```fare``` також є неперервною її треба закодувати до ```порядкової шкали``` 
Для цього скоритайтесь функцією ```pandas.qcut```.
[pandas.qcut](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.qcut.html)  
- необхідно перевірити, чи є вірною гіпотеза про те, що кількість виживших пасажирів для квартильних значеннь змінної ```fare``` відрізняється, та чи доцільно використовувати кодування на цю змінну(для подальшої роботи з закодованою фичею)

````python

data = pd.read_csv('titanic.csv')


data['Fare_bin'] = pd.qcut(data['Fare'], q=4, labels=False)


survived_by_fare_bin = data.groupby('Fare_bin')['Survived'].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=survived_by_fare_bin.index, y=survived_by_fare_bin.values)
plt.title('Відсоток виживших для кожного квартилю змінної Fare')
plt.xlabel('Квартиль змінної Fare')
plt.ylabel('Відсоток виживших')
plt.show()

print("Відсоток виживших для кожного квартилю змінної fare:")
print(survived_by_fare_bin)
````

Ваші висновки: гіпотеза про те, що кількість виживших пасажирів для квартильних значень змінної fare відрізняється, є вірною.

***Завдання 7:*** Необхідно створити нові змінні ```Family_size``` та ```Alone``` та проаналізувати їх. Змінна ```Family_size``` передає суму змінних ```Parch``` та ```SibSp```. Змінна ```Family_size``` буде відібражати комбіновані дані, які необхідні для перевірки гіпотези, що ```виживання пасажирів залежіть від кількості членів їх сім'ї```. Змінна ```Alone``` позначає наявніть сім'ї.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
sns.set_context('talk')

data = pd.read_csv('titanic.csv')

data['Family_size'] = data['Parch'] + data['SibSp']
data['Alone'] = (data['Family_size'] == 0).astype(int)

print(data['Family_size'].describe())
print(data['Alone'].value_counts())

plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Family_size', hue='Survived')
plt.title('Розподіл виживших за розміром сім\'ї')
plt.xlabel('Розмір сім\'ї')
plt.ylabel('Кількість')
plt.legend(title='Survived', labels=['Ні', 'Так'])
plt.xticks(rotation=45)
plt.show()
````

Ваші висновки: Середнє значення розміру сім'ї на борту лайнера становить приблизно 0.9, що може вказувати на те, що в середньому кожен пасажир мав менше одного брата/сестри або батьків/дітей на борту. За величиною квартилей можна зробити висновок, що більшість пасажирів подорожували без сім'ї (75% пасажирів мали розмір сім'ї 1 або менше). Аналіз змінної Alone показує, що понад половина пасажирів (близько 60%) подорожували без сім'ї.

***Завдання 11.*** Побудувати діаграми розподілу всіх числових змінних
